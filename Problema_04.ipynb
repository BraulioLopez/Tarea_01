Este programa incluye las funciones de una calculadora básica

```
def Suma(m, n):
  suma = m + n
  print(m, "+", n, "ES: ", suma)
def Resta (m, n):
  resta = m - n
  print(m, "-", n, "ES: ", resta)
def Multi (m, n):
  multi = m * n
  print(m, "x", n, "ES: ", multi)
def Divi (m, n):
  while n == 0:
    n = float(input("INGRESAR UN DIVISOR PERMITIDO: "))
  divi = m / n
  print(m, "/", n, "ES: ", divi)

print("A = SUMA    B = RESTA    C = MULTIPLICACIÓN    D = DIVISIÓN")
t = str(input("ELEGIR OPERACIÓN A REALIZAR: "))
if t == "a" or t =="A":
  print("SUMA")
  m = float(input("INSERTAR VALOR: "))
  n = float(input("INSERTAR VALOR: "))
  Suma (m, n)
elif t == "b" or t == "B":
  print("RESTA")
  m = float(input("INSERTAR VALOR: "))
  n = float(input("INSERTAR VALOR: "))
  Resta (m, n)
elif t == "c" or t =="C":
  print("MULTIPLICACIÓN")
  m = float(input("INSERTAR VALOR: "))
  n = float(input("INSERTAR VALOR: "))
  Multi (m, n)
elif t == "d" or t == "D":
  print("DIVISIÓN")
  m = float(input("INSERTAR DIVIDENDO: "))
  n = float(input("INSERTAR DIVISOR: "))
  Divi (m, n)
else:
  print("NO SE HA ELEGIDO UNA OPERACIÓN, REINICIAR PROGRAMA ")
```
