Este programa arroja el sueldo que se recibirá dependiendo el tiempo trabajado, arrojando aparte horarios predeterminados

```
def Sueldo(Horas, P=150):
  pago = Horas*P
  print("RECIBIRÁ: ", pago)
  día = P*8
  semana = P*40
  mes = P*160
  print("POR UN DÍA RECIBE: " + str(día))
  print("POR UNA SEMANA RECIBE: " + str(semana))
  print("POR UN MES RECIBE: " + str(mes))

Horas = int(input("HORAS TRABAJADAS: "))
Sueldo(Horas)
```
